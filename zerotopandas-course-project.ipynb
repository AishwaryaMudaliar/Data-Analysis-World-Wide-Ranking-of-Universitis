## Asking and Answering Questions

here are some interesting questions about the dataset that i came up with.



### Q1: How many institutions have a score above 90?

```python
above=dataset.loc[dataset['score']>90]
```

```python
above
```

we can see that total 8 instituions all the around the world have a score above 90.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(above,columns=['score','institution'])
plt.plot(df['score'],df['institution'],marker='o',color='maroon')
plt.xlabel('score', fontsize=20)
plt.ylabel('institutions', fontsize=20)
plt.show()
```

### Q2: How many institutions got ranked from USA


```python
usa=dataset.loc[dataset['country']=="USA"]
usa
```

There are a total of 230 institutions that got ranked belonged to USA 

### Q3: How many institutions got the 'quality of education' rank and  'quality of faculty' rank both above or equal to 10?

```python
bothranks=dataset.loc[(dataset['quality_of_education']<=10) & (dataset['quality_of_faculty']<=10)]
bothranks
```

It comes out that 8 institutions got both the ranks above 10

```python
import numpy as np
```

#### NumPy is a python library used for working with arrays. It also has functions for working in domain of linear algebra, fourier transform, and matrices.

```python
data= pd.DataFrame(bothranks, columns =['quality_of_education', 'quality_of_faculty'])
data
data.plot.bar() 
```

From the bar plot we can directly say that '4' the 'University of Oxford' has a higher rank in both quality of education and quality of faculty.

### Q4: What is the consistency between national rank and patents ?

```python
consistency=dataset[['national_rank','institution','patents']].set_index('institution').head(20)
consistency
```

```python
plt.title("consistency between national rank and patents")
sns.heatmap(consistency,fmt="d", cmap="icefire");
```

### Q5: How many institutions got national rank as 1 and have an influential rank under 50?

```python
nationalrank1=dataset.loc[(dataset['national_rank']==1) & (dataset['influence']<=50)]
nationalrank1
```

```python
sns.scatterplot(nationalrank1.influence,nationalrank1.institution,hue= nationalrank1.institution,s=100,palette='deep');
plt.xlabel('influence',fontsize=20)
plt.ylabel('institution',fontsize= 20)
plt.show();
```

There are a total of 6 institutions which got national rank as 1 and have influential rank under 50
